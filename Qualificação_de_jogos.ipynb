```python
import http.client
from bs4 import BeautifulSoup

# URL da página que contém as avaliações de jogos
url = 'https://www.metacritic.com/browse/games/score/metascore/all/all/filtered'

# Faz a solicitação HTTP
conn = http.client.HTTPSConnection("www.metacritic.com")
conn.request("GET", "/browse/games/score/metascore/all/all/filtered")
res = conn.getresponse()

# Verifica se a solicitação foi bem-sucedida
if res.status == 200:
    # Analisa o conteúdo da página com BeautifulSoup
    soup = BeautifulSoup(res.read(), 'html.parser')

    # Encontra os elementos HTML que contêm as informações desejadas
    game_containers = soup.find_all('td', class_='clamp-summary-wrap')

    # Itera sobre os elementos encontrados e extrai os dados
    for game in game_containers:
        game_title = game.find('h3').text.strip()
        game_score = game.find('div', class_='metascore_w').text.strip()

        # Imprime as informações
        print(f"Jogo: {game_title}")
        print(f"Avaliação: {game_score}")
        print()
else:
    print("Falha na solicitação HTTP.")

# Fecha a conexão
conn.close()
```
